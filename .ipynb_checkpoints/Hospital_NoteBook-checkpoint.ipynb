<h2> <font color="green">Question - 1 - Answer </font> </h2>

<i> Working only with the LOS_DAYS and AGE_AT_ADMISSION columns, provide an exploratory analysis of the dataset, extracting as much information as you can.</i>


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

nepjt_data = pd.read_csv("Non_elective_patient_journey_test.csv",index_col=0)
#pdDf = pd.DataFrame.from_csv("Non_elective_patient_journey_test.csv",index_col=0)

#plt.hist(pdDf["LOS_DAYS"], bins=5)
plt.hist(nepjt_data["LOS_DAYS"], bins=5)

plt.ylabel('Number of Patients')
plt.xlabel('LOS_DAYS')
plt.title('Non Elective Patient Journey LOS_DAYS for Patients')
plt.show()


#plt.hist(pdDf["AGE_AT_ADMISSION"], bins=5)
plt.hist(nepjt_data["AGE_AT_ADMISSION"], bins=5)

plt.ylabel('Number of Patients')
plt.xlabel('AGE_AT_ADMISSION')
plt.title('Non Elective Patient Journey AGE_AT_ADMISSION of Patients')
plt.show()
```

<h2> <font color="green">Questions - 2 - Answer </font> </h2>
Give the 10th, 18th and 91st percentiles for the LOS_DAYS distribution.

```python
losDaysData = np.array(nepjt_data["LOS_DAYS"])
ten = np.percentile(losDaysData, 10)

eighteen = np.percentile(losDaysData, 18)
nightyOne = np.percentile(losDaysData, 91)

print("10th Percentile fo LOS_DAYS = %s " % ten)
print("18th Percentile fo LOS_DAYS = %s " % eighteen)
print("91th Percentile fo LOS_DAYS = %s " % nightyOne)
```

<h2> <font color="green">Questions - 3 - Answer </font> </h2>

<li>a- Find the total, average and median length of stay for each day. </li>
<li>
b- Find the total, average and median length of stay by patient age, without the patients that entered during the month of April
 </li>

```python
total =  nepjt_data.groupby(['LOS_DAYS', 'DATE_OF_ADMISSION']).sum()
print (" Each Day Admissions Total %s " % total )


avg =  nepjt_data.groupby(['LOS_DAYS', 'DATE_OF_ADMISSION']).mean()
print (" Each Day Average %s " %avg )


med =  nepjt_data.groupby(['LOS_DAYS', 'DATE_OF_ADMISSION']).median()
print (" Each Day Median Total %s " %med )


total1 =  nepjt_data.groupby(['AGE_AT_ADMISSION', 'DATE_OF_ADMISSION']).sum()
print (" Admissions Total by AGE %s " % total1 )


avg1 =  nepjt_data.groupby(['AGE_AT_ADMISSION', 'DATE_OF_ADMISSION']).mean()
print (" Admissions Average by AGE %s " %avg1 )


med1 =  nepjt_data.groupby(['AGE_AT_ADMISSION', 'DATE_OF_ADMISSION']).median()
print (" Admissions Median by AGE %s " %med1 )



```
